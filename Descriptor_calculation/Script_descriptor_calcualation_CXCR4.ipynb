# Script for loading of dataset in jupyter notebook

```python
import pandas as pd
import pickle
from rdkit import Chem
from rdkit.Chem import Descriptors, rdMolDescriptors, PandasTools, AllChem
```

```python
Location = r'C:\Users\wanih\\CXCR4\Dataset_external.csv'
df = pd.read_csv(Location)
```

# Script for calculating decriptors

```python
# List of all available descriptors in RDKit
descriptor_names = [desc[0] for desc in Descriptors._descList]
# Function to calculate descriptors for a given molecule (SMILES string)
def calculate_descriptors(smiles):
    mol = Chem.MolFromSmiles(smiles)
    descriptor_values = {}
    
    if mol is not None:
        # Calculate regular descriptors using RDKit Descriptors module
        for desc_name in descriptor_names:
            # Skip fingerprint-related descriptors like "Fingerprint" or "MACCS"
            if "Fingerprint" not in desc_name and "MACCS" not in desc_name:
                func = getattr(Descriptors, desc_name)
                try:
                    # Calculate the descriptor value for the molecule
                    descriptor_values[desc_name] = func(mol)
                except Exception as e:
                    # Handle any error (e.g., descriptor not calculable)
                    descriptor_values[desc_name] = None
    else:
        # Handle invalid SMILES or empty molecule
        for desc_name in descriptor_names:
            descriptor_values[desc_name] = None
        
    return descriptor_values
# Apply the descriptor calculation for each molecule in the 'Canonical_Smiles' column
descriptor_columns = df['Smiles'].apply(calculate_descriptors)
# Convert the result (which is a list of dictionaries) into a DataFrame
descriptors_data = pd.DataFrame(descriptor_columns.tolist())
data_final = pd.concat([df,descriptors_data],axis=1)
```
